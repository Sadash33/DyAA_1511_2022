```python
arch = open("Quijote.txt.txt", 'rt')
texto = arch.read()

def string_match_brute(A, P):
  contador = 0
  for i in range(len(A) - len(P) + 1):
    for j in range(len(P)):
      if P[j] == A[i + j]:
        pass
      else:
        break
    if j + 1 == len(P) and A[i + j] == P[j]:
      contador = contador + 1
      print(f"Se encontró en la posición {i}")
  print(f"La palabra {P} existe {contador} veces en el texto")
  return contador

contador = []
contador.append(string_match_brute(texto, "molino"))
contador.append(string_match_brute(texto, "hidalgo"))
contador.append(string_match_brute(texto, "merced"))
arch.close()

print("Hidalgo: 46 veces ")
print("Molino: 5 veces")
print("Merced: 590 veces ")
```
