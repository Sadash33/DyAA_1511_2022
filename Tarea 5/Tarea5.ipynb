```python
cantidad = float(input("Ingrese la cantidad: ")) #O(1) O(1)
denominacion =[200, 100, 20, 10, 5, 1, 0.5] #O(1) O(7)
lista =[0, 0, 0, 0, 0, 0, 0] #O(1) O(7)
 
 
for i in range(0, len(denominacion)): #O(7) 
    cambio = denominacion[i]
    while cantidad - cambio >= 0: #O(7) 
        lista[i] += 1
        cantidad -= cambio
 
for i in range(0, len(lista)): #O(7) 
    if lista[i] > 0:
        print(f"{lista[i]} billetes/monedas de {denominacion[i]}")
```
